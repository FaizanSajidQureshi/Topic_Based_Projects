# Step 2: Feature Engineering

```python
# Creating New Features
print("FEATURE ENGINEERING")

# 1. Feature for Balance change
print("Balance change feature")
fraud_data['balanceChangeOrig'] = fraud_data['newbalanceOrig'] - fraud_data['oldbalanceOrg']
fraud_data['balanceChangeDest'] = fraud_data['newbalanceDest'] - fraud_data['oldbalanceDest']

# 2. Amount Relative to Balance
print("Amount ratio feature")
fraud_data['AmountTOBalanceRatio'] = fraud_data['amount'] / (fraud_data['oldbalanceOrg'] + 1)

# 3. Merchant Destination Flag
print("Merchant Flag")
fraud_data['MerchantDest'] = fraud_data['nameDest'].str.startswith('M').astype(int)

# 4. Balance Error Detection
print("Balance error feature")
fraud_data['BalanceErrorOrig'] = fraud_data['newbalanceOrig'] + fraud_data['amount'] - fraud_data['oldbalanceOrg']

# 5. Transaction Size Categories
print("Transaction size categories")
bins = [0, 100, 1000, 10000, float('inf')]
labels = ['Small', 'Medium', 'Large', 'X_Large']
fraud_data['transactionSize'] = pd.cut(fraud_data['amount'], bins=bins, labels=labels)

# 6. Transaction Typewise Flagging
print("Transaction type flags")
fraud_data['isTransfer'] = (fraud_data['type'] == 'TRANSFER').astype(int)
fraud_data['isCashOut'] = (fraud_data['type'] == 'CASH_OUT').astype(int)
fraud_data['isPayment'] = (fraud_data['type'] == 'PAYMENT').astype(int)

# Defining function for recalling feature information
def print_feature_info(df, features, feature_name):
    print(f"\n{feature_name}:")
    print("First 5 rows:")
    print(df[features].head())
    print("\nSummary Stats:")
    print(df[features].describe())
    print(f"Total features in dataset: {df.shape[1]}")

# Display of new features
new_features = [
    'balanceChangeOrig', 'balanceChangeDest', 'AmountTOBalanceRatio',
    'MerchantDest', 'BalanceErrorOrig', 'transactionSize',
    'isTransfer', 'isCashOut', 'isPayment'
]

print_feature_info(fraud_data, new_features, "NEW FEATURES CREATED")
print("Feature engineering completed succesfully.")
```

# Step 3: Preprocessing

```python
fraud_data = pd.read_csv(FRAUD_DATA).dropna(how="all")

fraud_data['balanceChangeOrig'] = fraud_data['newbalanceOrig'] - fraud_data['oldbalanceOrg']
fraud_data['balanceChangeDest'] = fraud_data['newbalanceDest'] - fraud_data['oldbalanceDest']
fraud_data['AmountTOBalanceRatio'] = fraud_data['amount'] / (fraud_data['oldbalanceOrg'] + 1)
fraud_data['MerchantDest'] = fraud_data['nameDest'].str.startswith('M').astype(int)
fraud_data['BalanceErrorOrig'] = fraud_data['newbalanceOrig'] + fraud_data['amount'] - fraud_data['oldbalanceOrg']

# Transaction Size
bins = [0, 100, 1000, 10000, float('inf')]
labels = ['Small', 'Medium', 'Large', 'X_Large']
fraud_data['transactionSize'] = pd.cut(fraud_data['amount'], bins=bins, labels=labels)

print("Encoding categorical features...")

# One-hot encoding for 'type'
type_dummies = pd.get_dummies(fraud_data['type'], prefix='type', dtype=int)

# Label encoding for transaction size
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
fraud_data['encoded_transactionSize'] = le.fit_transform(fraud_data['transactionSize'])

# Drop unused/redundant columns
fraud_data.drop(
    columns=['type', 'transactionSize', 'nameOrig', 'nameDest', 'isTransfer', 'isCashOut', 'isPayment'],
    inplace=True, errors='ignore'
)

# Add dummy columns
fraud_data = pd.concat([fraud_data, type_dummies], axis=1)

# Convert all columns to numeric
fraud_data = fraud_data.apply(pd.to_numeric, errors='coerce')

print(f" Dataset shape after encoding: {fraud_data.shape}")
print(" All columns numeric:", fraud_data.select_dtypes(include='object').empty)

encoded_features = [col for col in fraud_data.columns if col.startswith('type_') or col == 'encoded_transactionSize']
print("\nEncoded feature sample:")
print(fraud_data[encoded_features].head())
```

```python
# Train_Test_Split
# Select all columns except 'isFraud' and 'day' for features X
X = fraud_data.drop(columns=['isFraud', 'day'], errors='ignore')
y = fraud_data['isFraud']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

print("Data Splitting Completed!!")
```

```python
# Feature Scaling
# 1. Standard Scaler Method
scaler_S = StandardScaler()
X_train_standard = scaler_S.fit_transform(X_train)
X_test_standard = scaler_S.transform(X_test)
print("Standard Scaling Completed")

# 2. Min-Max Scaler Method
scaler_M = MinMaxScaler()
X_train_mm = scaler_M.fit_transform(X_train)
X_test_mm = scaler_M.transform(X_test)
print("MinMax Scaling Completed")

# 3. Robust Scaler Method
from sklearn.preprocessing import RobustScaler
scaler_R = RobustScaler()
X_train_r = scaler_R.fit_transform(X_train)
X_test_r = scaler_R.transform(X_test)
print("Robust Scaling Completed")

# Comparison of scaling results
print("\nScaling comparison for 'amount' feature:")
print("Original values:", X_train['amount'].head().values)
print("Standard Scaled:", X_train_standard[:3, 0])
print("MinMax Scaled:", X_train_mm[:3, 0])

print("\nFeature Scaling Completed!!")

print("\nScaled fraud_datasets:")
print("~ Standard Scaler (mean=0, std=1)")
print("~ MinMax Scaler (range 0-1)")
print("~ Robust Scaler (robust to outliers)")
```

# Step 4: Model Evaluation

```python
print("Available variables:")
print([var for var in dir() if 'train' in var.lower() or 'test' in var.lower()])
```

```python
# Model Evaluation
print("MODEL TRAINING AND EVALUATION")

models = {
    'Logistic Regression': LogisticRegression(random_state=42),
    'Decision Tree': DecisionTreeClassifier(random_state=42),
    'Random Forest': RandomForestClassifier(random_state=42),
    'Gradient Boosting': GradientBoostingClassifier(random_state=42),
    'XGBoost': XGBClassifier(random_state=42, eval_metric='logloss'),
    'AdaBoost': AdaBoostClassifier(random_state=42),
    'SVM': SVC(random_state=42, probability=True)
}
results = {}

print("Training and evaluating models")
print("=" * 80)

for name, model in models.items():
    print(f"\n {name}")

    # Training the model
    if name in ['Logistic Regression', 'SVM']: # For Scaled Data
        model.fit(X_train_standard, y_train)
        y_pred = model.predict(X_test_standard)
        y_pred_proba = model.predict_proba(X_test_standard)[:, 1]
    else:                                      # For Non-Scaled Data
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)
        y_pred_proba = model.predict_proba(X_test)[:, 1]

    # Calculation of metrics
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    roc_auc = roc_auc_score(y_test, y_pred_proba)

    # Storing the results
    results[name] = {
        'accuracy': accuracy,
        'precision': precision,
        'recall': recall,
        'f1_score': f1,
        'roc_auc_score': roc_auc,
        'model': model
    }

    # Printing Results
    print(f"Accuracy:  {accuracy:.4f}")
    print(f"Precision:  {precision:.4f}")
    print(f"Recall:  {recall:.4f}")
    print(f"F1_Score:  {f1:.4f}")
    print(f"ROC_AUC_Score:  {roc_auc:.4f}")

print("\n" + "=" * 80)
print("Model Evaluation Completed!!")
```

# Step 5: Clustering Model Building

```python
print("FINDING OPTIMAL NUMBER OF CLUSTERS")

# Elbow method for K-means
inertias = []
silhouette_scores = []
k_range = range(2, 11)

for k in k_range:
  try:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X_scaled)
    inertias.append(kmeans.inertia_)

    # Silhouette score
    if k > 1:
      score = silhouette_score(X_scaled, kmeans.labels_)
      silhouette_scores.append(score)
  except Exception as e:
    print(f" Error with k={k}: {e}")
    inertias.append(0)
    silhouette_scores.append(0)

# Plot elbow curve and silhouette scores
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))

# Elbow curve
ax1.plot(k_range, inertias, marker='o')
ax1.set_xlabel('Number of Clusters')
ax1.set_ylabel('Inertia')
ax1.set_title('Elbow Method for Optimal K')
ax1.grid(True)

# Silhouette scores
ax2.plot(range(2, 11), silhouette_scores, marker='o', color='orange')
ax2.set_xlabel('Number of Clusters')
ax2.set_ylabel('Silhouette Score')
ax2.set_title('Silhouette Scores for Different K')
ax2.grid(True)

plt.tight_layout()
plt.show()

# Find optimal k based on silhouette score
optimal_k = range(2, 11)[np.argmax(silhouette_scores)]
print(f"Optimal number of clusters based on silhouette score: {optimal_k}")
```

# Step 6: Model Evaluation

```python
def apply_clustering_models(X_scaled, X_pca, optimal_k):
    print("APPLYING CLUSTERING MODELS")

    # Validate data before clustering
    if np.any(np.isnan(X_scaled)) or np.any(np.isinf(X_scaled)):
        print(" CRITICAL: Cleaning data before clustering")
        X_scaled = np.nan_to_num(X_scaled, nan=0.0, posinf=1e6, neginf=-1e6)

    # K-means Clustering
    print("Applying K-means clustering")
    try:
        kmeans = KMeans(n_clusters=optimal_k, random_state=42, n_init=15)
        kmeans_labels = kmeans.fit_predict(X_scaled)
        kmeans_labels_pca = kmeans.fit_predict(X_pca[:, :2])  # Using first 2 PCA components
    except Exception as e:
        print(f" K-means failed: {e}")
        # Fallback: use random labels
        kmeans_labels = np.random.randint(0, optimal_k, len(X_scaled))
        kmeans_labels_pca = kmeans_labels

    # Agglomerative Clustering
    print("Applying Agglomerative clustering")
    try:
        agglo = AgglomerativeClustering(n_clusters=optimal_k)
        agglo_labels = agglo.fit_predict(X_scaled)
    except Exception as e:
        print(f" Agglomerative failed: {e}")
        # Fallback: use random labels
        agglo_labels = np.random.randint(0, optimal_k, len(X_scaled))

    # Evaluate clustering performance
    def evaluate_clustering(X, labels, model_name):
        try:
            if len(np.unique(labels)) > 1:  # Need at least 2 clusters for these metrics
                silhouette = silhouette_score(X, labels)
                calinski = calinski_harabasz_score(X, labels)
                davies = davies_bouldin_score(X, labels)

                print(f"{model_name} Performance:")
                print(f"  Silhouette Score: {silhouette:.3f}")
                print(f"  Calinski-Harabasz Score: {calinski:.3f}")
                print(f"  Davies-Bouldin Score: {davies:.3f}")
                return silhouette, calinski, davies
        except Exception as e:
            print(f" Evaluation failed for {model_name}: {e}")
        return None, None, None

    print("\nClustering Performance Evaluation:")
    kmeans_scores = evaluate_clustering(X_scaled, kmeans_labels, "K-means")
    agglo_scores = evaluate_clustering(X_scaled, agglo_labels, "Agglomerative")

    return kmeans_labels, kmeans_labels_pca, agglo_labels, kmeans


kmeans_labels, kmeans_labels_pca, agglo_labels, kmeans_model = apply_clustering_models(X_scaled, X_pca, optimal_k)
```

```python
def hybrid_features(data):
  supervised_features = [
      'amount', 'oldbalanceOrg', 'newbalanceOrig', 'oldbalanceDest', 'newbalanceDest', 'balanceChangeOrig', 'balanceChangeDest',
      'AmountTOBalanceRatio', 'BalanceErrorOrig', 'MerchantDest', 'isTransfer', 'isCashOut', 'isPayment', 'type_CASH_IN',
        'type_CASH_OUT', 'type_DEBIT', 'type_PAYMENT', 'type_TRANSFER', 'encoded_transactionSize'
  ]
  Unsupervised_features = [
      'BALANCE', 'BALANCE_FREQUENCY', 'PURCHASES', 'ONEOFF_PURCHASES', 'INSTALLMENTS_PURCHASES', 'CASH_ADVANCE', 'PURCHASES_FREQUENCY',
      'ONEOFF_PURCHASES_FREQUENCY', 'CASH_ADVANCE_FREQUENCY', 'CASH_ADVANCE_TRX', 'PURCHASES_TRX', 'CREDIT_LIMIT', 'PAYMENTS',
      'MINIMUM_PAYMENTS', 'PRC_FULL_PAYMENT', 'TOTAL_SPENDING', 'CREDIT_UTILIZATION', 'PAYMENT_TO_BALANCE_RATIO'
  ]

  avb_sup = [f for f in supervised_features if f in data.columns]
  avb_unsup = [f for f in Unsupervised_features if f in data.columns]

  X_hybrid = data[avb_sup + avb_unsup].copy()

  return X_hybrid
```

# Step 2: MTL Model Architecture

```python
class MTL_Fraud_Segmenter(Model):
  def __init__(self, input_dim, num_clusters=4):
    super(MTL_Fraud_Segmenter, self).__init__()

    # Shared Encoders
    self.shared_dense_1 = layers.dense(128, activation='relu')
    self.shared_dense_2 = layers.dense(64, activation='relu')
    self.shared_dense_3 = layers.dense(32, activation='relu')

    # Task 1: Fraud Detection
    self.fraud_hd = layers.dense(16, activation='relu')
    self.fraud_output = layers.dense(1, activation='sigmoid', name= 'fraud_detection')

    # Task 2: Customer Segmentation
    self.embedding_hd = layers.dense(16, activation='relu')
    self.embedding_output = layers.Dense(8, activation='linear', name='customer_segmentation')

  def call(self, inputs):
    # Shared Features Extraction
    x = self.shared_dense_1(inputs)
    x = self.shared_dense_2(x)
    shared_features = self.shared_dense_3(x)

    # Heads & Outputs
    fraud_features = self.fraud_hd(shared_features)
    fraud_pred = self.fraud_output(fraud_features)
    embedding_features = self.embedding_hd(shared_features)
    customer_segmentation = self.embedding_output(embedding_features)

    return {'fraud_detection': fraud_pred, 'customer_segmentation':customer_segmentation}
```

# Step 3: Model Training

```python
def train_busted_ai(X_train, y_fraud_train):
  X_scaled = scaler.fit_transform(X_train)
  model = MTL_Fraud_Segmenter(input_dim=X_scaled.shape[1])
  def mtl_loss(fraud_labels, embeddings, alpha=0.7):
    # Binary Crossentrophy
    fraud_loss = tf.keras.losses.binary_crossentropy(fraud_labels[0], fraud_labels[1])

    # Maximise Separation for Clustering
    embedding_loss = -tf.reduce_mean(
        tf.norm(embeddings - tf.reduce_mean(embeddings, axis=0), axis=1)

    )
    return alpha * fraud_loss + (1 - alpha) * embedding_loss

    # Model Compilation
    model.compile(optimizer='adam', loss={'fraud_detection': 'binary_crossentropy', 'customer_segmentation': mtl_loss},
                  loss_weights = {'fraud_detection': 0.7, 'customer_segmentation': 0.3},
                  metrics={'fraud_detection': ['accuracy', 'precision', 'recall']}
                  )

    # Model Training
    history = model.fit(X_scaled, {'fraud_detection': y_fraud_train,
                                   'customer_segmentation': X_scaled},
                        epochs=50,
                        batch_size=32,
                        validation_split=0.2,
                        verbose=1
                        )
    return model, history
```

# Step 4: Model Prediction

```python
def predict_busted_ai(model, X_data):
  fraud_pred, customer_segmentation = model.predict(X_data)

  # Apply Clusterings on embeddings
  from sklearn.cluster import KMeans
  kmeans = KMeans(n_clusters=4, random_state=42)
  customer_segments = kmeans.fit_predict(customer_segmentation)

  return {
      'fraud_pred': fraud_pred,
      'customer_segmentation': customer_segmentation,
      'customer_segments': customer_segments,
      'fraud_scores': fraud_pred.flatten()
  }
```

# Step 5: Model Evaluation

```python
def evaluate_busted_ai(results, y_true_fraud):
  fraud_metrics = {
      'accuracy': accuracy_score(y_true_fraud, results['fraud_pred'] > 0.5),
      'precision': precision_score(y_true_fraud, results['fraud_pred'] > 0.5),
      'recall': recall_score(y_true_fraud, results['fraud_pred'] > 0.5),
      'f1_score': f1_score(y_true_fraud, results['fraud_pred'] > 0.5),
      'roc_auc': roc_auc_score(y_true_fraud, results['fraud_pred'])
  }

  clustering_metrics = {
      'silhouette_score': silhouette_score(results['customer_segmentation'], results['customer_segments']),
      'cluster_quality': len(np.unique(results['customer_segments']))
  }

  segment_fraud_rates = []
  for segment in np.unique(results['customer_segments']):
    segment_mask = results['customer_segments'] == segment
    fraud_rate = np.mean(results['fraud_scores'][segment_mask])
    segment_fraud_rates.append(fraud_rate)

  return {
      'fraud_metrics': fraud_metrics,
      'clustering_metrics': clustering_metrics,
      'segment_fraud_rates': np.corrcoef(segment_fraud_rates, range(len(segment_fraud_rates)))[0,1]
  }
```

# Step 6: Neural Network Implementation

```python
class busted_ai(Model):
  def __init__(self, input_dim, num_clusters=4):
    super(busted_ai, self).__init__()

    self.shared_dense_1 = layers.Dense(256, activation='relu', kernel_regularizer = tf.keras.regularizers.l2(0.001))
    self.dropout_1 = layers.Dropout(0.3)
    self.shared_dense_2 = layers.Dense(128, activation='relu', kernel_regularizer=tf.keras.regularizers.l2(0.001))
    self.shared_dense_3 = layers.Dense(64, activation='relu')
    self.fraud_dense_1 = layers.Dense(32, activation='relu')
    self.fraud_dense_2 = layers.Dense(16, activation='relu')
    self.fraud_output = layers.Dense(1, activation='sigmoid', name='fraud_detection')

    self.embedding_dense_1 = layers.Dense(32, activation='relu')
    self.embedding_dense_2 = layers.Dense(16, activation='relu')
    self.embedding_output = layers.Dense(8, activation='linear', name='customer_segmentation')

  def call(self, inputs, training=False):
    x = self.shared_dense_1(inputs)
    x = self.dropout_1(x, training=training)
    x = self.shared_dense_2(x)
    x = self.dropout_1(x, training=training)
    shared_features = self.shared_dense_3(x)

    fraud_x = self.fraud_dense_1(shared_features)
    fraud_x = self.fraud_dense_2(fraud_x)
    fraud_pred = self.fraud_output(fraud_x)

    embedding_x = self.embedding_dense_1(shared_features)
    embedding_x = self.embedding_dense_2(embedding_x)
    customer_segmentation = self.embedding_output(embedding_x)

    return fraud_pred, customer_segmentation

  def get_shared_features(self, inputs):
    x = self.shared_dense_1(inputs)
    x = self.shared_dense_2(x)
    return self.shared_dense_3(x)
```

# Step 7: Busted_AI Model Training Setup

```python
def prepare_ai_data():
  fraud_features = [
      'amount', 'oldbalanceOrg', 'newbalanceOrig', 'oldbalanceDest', 'newbalanceDest',
      'balanceChangeOrig', 'balanceChangeDest', 'AmountTOBalanceRatio', 'BalanceErrorOrig',
      'MerchantDest', 'isTransfer', 'isCashOut', 'isPayment', 'type_CASH_IN', 'type_CASH_OUT',
      'type_DEBIT', 'type_PAYMENT', 'type_TRANSFER', 'encoded_transactionSize'
  ]

  segmentation_features = [
      'BALANCE', 'BALANCE_FREQUENCY', 'PURCHASES', 'ONEOFF_PURCHASES',
      'INSTALLMENTS_PURCHASES', 'CASH_ADVANCE', 'PURCHASES_FREQUENCY',
      'ONEOFF_PURCHASES_FREQUENCY', 'CASH_ADVANCE_FREQUENCY', 'CASH_ADVANCE_TRX',
      'PURCHASES_TRX', 'CREDIT_LIMIT', 'PAYMENTS', 'MINIMUM_PAYMENTS',
      'PRC_FULL_PAYMENT', 'TOTAL_SPENDING', 'CREDIT_UTILIZATION', 'PAYMENT_TO_BALANCE_RATIO'
  ]

  avb_fraud = [f for f in fraud_features if f in fraud_data.columns]
  avb_seg = [f for f in segmentation_features if f in seg_data.columns]
  hybrid_features = avb_fraud + avb_seg
  X_hybrid = fraud_data[avb_fraud].copy()

  X_hybrid = X_hybrid.fillna(X_hybrid.median())
  y_fraud = fraud_data['isFraud'].values

  return X_hybrid, y_fraud

print("Preparing data")
X_hybrid, y_fraud = prepare_ai_data()

# Data Splitting
X_train, X_test, y_train, y_test = train_test_split(
    X_hybrid, y_fraud, test_size=0.3, random_state=42, stratify=y_fraud
)

# Feature Scaling
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

with open("scaler.pkl", "wb") as f:
    pickle.dump(scaler, f)
print("Scaler saved as scaler.pkl")

print(f"Hybrid feature matrix shape: {X_train_scaled.shape}")
print(f"Fraud cases in training: {np.sum(y_train)} ({np.mean(y_train)*100:.2f}%)")
```

# Step 8: Busted_AI Model Training

```python
def create_ai_model(input_dim):
    inputs = tf.keras.Input(shape=(input_dim,))

    # Shared layers
    x = layers.Dense(256, activation='relu', kernel_regularizer=tf.keras.regularizers.l2(0.001))(inputs)
    x = layers.Dropout(0.3)(x)
    x = layers.Dense(128, activation='relu', kernel_regularizer=tf.keras.regularizers.l2(0.001))(x)
    x = layers.Dropout(0.3)(x)
    shared_features = layers.Dense(64, activation='relu')(x)

    # Fraud detection branch
    fraud_x = layers.Dense(32, activation='relu')(shared_features)
    fraud_x = layers.Dense(16, activation='relu')(fraud_x)
    fraud_output = layers.Dense(1, activation='sigmoid', name='fraud_detection')(fraud_x)

    # Customer segmentation branch
    embedding_x = layers.Dense(32, activation='relu')(shared_features)
    embedding_x = layers.Dense(16, activation='relu')(embedding_x)
    segmentation_output = layers.Dense(8, activation='linear', name='customer_segmentation')(embedding_x)

    # Create model
    model = tf.keras.Model(inputs=inputs, outputs=[fraud_output, segmentation_output])

    # Compile model
    model.compile(
        optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
        loss=
            {
            'fraud_detection': tf.keras.losses.BinaryCrossentropy(),
            'customer_segmentation': tf.keras.losses.MeanSquaredError()
        },
        loss_weights=
            {
            'fraud_detection': 0.8,
            'customer_segmentation': 0.2
        },
        metrics=
            {
            'fraud_detection': ['accuracy', 'precision', 'recall'],
        }
    )
    return model

# Create and Train Model
print("Creating Busted AI Model...")
busted_ai_model = create_ai_model(input_dim=X_train_scaled.shape[1])

# Define callbacks list correctly
callbacks = [
    tf.keras.callbacks.EarlyStopping(patience=10, restore_best_weights=True),
    tf.keras.callbacks.ReduceLROnPlateau(factor=0.5, patience=5),
    tf.keras.callbacks.ModelCheckpoint('busted_ai_model.h5', save_best_only=True)
]

print("Training Busted AI Model...")
history = busted_ai_model.fit(
    X_train_scaled,
    {
        'fraud_detection': y_train,
        'customer_segmentation': np.zeros((len(y_train), 8)) # Placeholder for segmentation embeddings
    },
    epochs=20,
    batch_size=64,
    validation_split=0.2,
    callbacks=callbacks,
    verbose=1
)
```

# Step 9: Busted_AI Model Evaluation

```python
def evaluate_ai(model, X_test, y_test):
  # Get Predictions
  fraud_pred_proba, customer_segmentation = model.predict(X_test, verbose=0)
  fraud_pred_proba = fraud_pred_proba.flatten()
  fraud_pred = (fraud_pred_proba > 0.5).astype(int)

  # Apply clustering on emmbeddings
  from sklearn.cluster import KMeans
  kmeans = KMeans(n_clusters=4, random_state=42)

  try:
    customer_segments = kmeans.fit_predict(customer_segmentation)
    n_clusters_found = len(np.unique(customer_segments))

    # If clustering failed, use random segments for now
    if n_clusters_found < 2:
      print(f"Clustering found only {n_clusters_found} cluster(s). Using random segments.")
      customer_segments = np.random.randint(0, 4, len(customer_segmentation))

  except Exception as e:
    print(f"Clustering failed: {e}. Using random segments.")
    customer_segments = np.random.randint(0, 4, len(customer_segmentation))

  # Fraud Detection Metrics
  fraud_metrics = {
      'accuracy': accuracy_score(y_test, fraud_pred),
      'precision': precision_score(y_test, fraud_pred),
      'recall': recall_score(y_test, fraud_pred),
      'f1_score': f1_score(y_test, fraud_pred),
      'roc_auc': roc_auc_score(y_test, fraud_pred_proba),
      'confusion_matrix': confusion_matrix(y_test, fraud_pred)
  }

  # Clustering Metrics
  clustering_metrics = {
      'silhouette_score': silhouette_score(customer_segmentation, customer_segments),
      'calinski_harabasz_score': calinski_harabasz_score(customer_segmentation, customer_segments),
      'davies_bouldin_score': davies_bouldin_score(customer_segmentation, customer_segments),
      'n_clusters': len(np.unique(customer_segments))
  }

  # Cross Task Analysis
  segment_analysis = {}
  for segment_id in np.unique(customer_segments):
    segment_mask = customer_segments == segment_id
    segment_fraud_rate = np.mean(fraud_pred_proba[segment_mask])
    segment_avg_fraud_score = np.mean(fraud_pred_proba[segment_mask])
    segment_analysis[segment_id] = {
        'size': np.sum(segment_mask),
        'fraud_rate': segment_fraud_rate,
        'avg_fraud_score': segment_avg_fraud_score,
        'high_risk': segment_avg_fraud_score > np.mean(y_test)
    }

  return {
      'fraud_predictions': fraud_pred,
      'fraud_probabilities': fraud_pred_proba,
      'customer_segmentation': customer_segmentation,
      'customer_segments': customer_segments,
      'fraud_metrics': fraud_metrics,
      'clustering_metrics': clustering_metrics,
      'segment_analysis': segment_analysis
  }

# Model Evaluation
print("Evaluating Busted AI Model")
results = evaluate_ai(busted_ai_model, X_test_scaled, y_test)

# Print Results
print("Busted AI Model Results")
print("-" * 60)

print("Fraud Detection Performance:")
for metric, value in results['fraud_metrics'].items():
  if metric != 'confusion_matrix':
    print(f"{metric:15}: {value:.4f}")

print("\nClustering Performance:")
for metric, value in results['clustering_metrics'].items():
  print(f"{metric:25}: {value:.4f}")

print("Cross-Task Analysis:")
for segment_id, analysis in results['segment_analysis'].items():
  risk_level = "High Risk" if analysis['high_risk'] else "Low Risk"
print(f"Segment {segment_id}: {analysis['size']:4} customers |"
f"Fraud rate: {analysis['fraud_rate']:.3f} | {risk_level}")
```
